* ‰∏ãÂçäÈÉ®ÔºöÁî±ÂÄíÁΩÆÁöÑÂúìÂºßÂíåÂπ≥ÁßªÂæåÁöÑÂúìÂºßÁµÑÊàêÔºåÂΩ¢ÊàêËùôËù†Ë∫´È´î„ÄÇ

```python
import numpy as np
import matplotlib.pyplot as plt

def batman_curve():
    x = np.linspace(-7, 7, 400)

# ÂÆöÁæ©ÂàÜÊÆµÂáΩÊï∏ÔºåÂú®‰∏çÂêåÁØÑÂúçÂÖß‰ΩøÁî®‰∏çÂêåÁöÑÂáΩÊï∏
# |x|<3Ôºåy=2.25*sqrt(1‚àí(7x‚Äã)2)‚Äã ÂçäÂúìÂΩ¢
# 3<=|x|<=7Ôºåy=‚Äã|x/2|‚Äã‚Äã‚àí0.75

    y1 = np.piecewise(x,
        [np.abs(x) < 3, (np.abs(x) >= 3) & (np.abs(x) < 7)],
        [lambda x: 2.25*np.sqrt(1 - (x/7)**2),
         lambda x: np.abs(x/2) - 0.75])

#‰∏ãÂçäÈÉ®Êõ≤Á∑ö

    y2 = np.piecewise(x,
        [np.abs(x) < 4, (np.abs(x) >= 4) & (np.abs(x) < 7)],
        [lambda x: -3*np.sqrt(1 - (x/7)**2),
         lambda x: -3*np.sqrt(1 - (x/7)**2) - 0.25])

    return x, y1, y2

x, y1, y2 = batman_curve()

#Áï´Âúñ
plt.figure(figsize=(8,6))
plt.plot(x, y1, 'k', linewidth=2)
plt.plot(x, y2, 'k', linewidth=2)


plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)


plt.xlim(-8, 8)
plt.ylim(-5, 5)

plt.title("Batman", fontsize=14, fontweight='bold')

plt.show()
```
